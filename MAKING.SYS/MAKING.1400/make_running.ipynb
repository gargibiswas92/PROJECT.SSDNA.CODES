```python
inp1 = open("dna1.pdb", "r")

atom = []
atom_num = []
atom_name = []
chain_1 = []
chain_2 = []
res_num = []
x_cord = []
y_cord = []
z_cord = []
occu = []
b_fact = []
atom_type = []

for lines in inp1:
    if lines.startswith("ATOM"):
        arr = lines.split()
        atom.append(str(lines[0:4]))
        atom_num.append(int(lines[4:11]))
        atom_name.append(str((lines[11:15]).strip()))
        chain_1.append(str((lines[15:19]).strip()))
        chain_2.append(str((lines[19:21]).strip()))
        res_num.append(int(lines[21:26]))
        x_cord.append(float(lines[26:38]))
        y_cord.append(float(lines[38:46]))
        z_cord.append(float(lines[46:54]))
        occu.append(float(lines[54:60]))
        b_fact.append(float(lines[60:66]))
        #atom_type.append(str(arr[11]))

for i in range(len(atom_num)):
    if atom_num[i] == 6039:
        a = x_cord[i]
        b = y_cord[i]
        c = z_cord[i]
        #print(a, b, c)
    if atom_num[i] == 10865:
        m = x_cord[i]
        n = y_cord[i]
        l = z_cord[i]
aa = a - m
bb = b - n
cc = c - l
    
x_mod = []
y_mod = []
z_mod = []

out1 = open("mod_dna1.pdb", "w")

for j in range(len(atom_num)):
    p = x_cord[j] + aa
    q = y_cord[j] + bb
    r = z_cord[j] + cc
    x_mod.append(p)
    y_mod.append(q)
    z_mod.append(r)
    
    #ss = '{}{:>7s}{}{:<3s}{}{:>4s}{:>12.2f}{:>8.2f}{:>8.2f}{}{}{}'.format('ATOM', str(atom_num[j]), str(atom_name[j]),'   C D ', str(res_num[j]), x_mod[j], y_mod[j], z_mod[j], '  1.00', '  0.00', "\n")
    ss = '{}{:>7s}{}{:<3s}{}{:>4s}{:>12.3f}{:>8.3f}{:>8.3f}{}{}{}'.format('ATOM', str(atom_num[j]), ' ', str(atom_name[j]),'   C D ', str(res_num[j]), x_mod[j], y_mod[j], z_mod[j], '  1.00', '  0.00', "\n")
    out1.writelines(ss)
    
inp1.close()
out1.close()
```
